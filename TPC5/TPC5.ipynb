```python

def criar_turma():
    turma=[]
    return turma

def inserir_aluno(turma):
    id = input("Introduza o id do aluno")
    nome = input("Insira o nome do aluno.")
    notaTPC = float(input("Insira a nota do TPC."))
    notaProj = float(input("Insira a nota do projeto."))
    notaTeste = float(input("Insira a nota do teste."))
    b = [notaTPC, notaProj, notaTeste]
    a = (nome, id, b)
    turma.append(a)
    return turma

def listar_turma(turma):
    for a in turma:
            print(f"ID: {a[1]} | Nome: {a[0]} | Notas: {a[2]}")

def consultar_id(turma,i):
    for a in turma:
        if i == a[1]:
            print(f"ID: {a[1]} | Nome: {a[0]} | Notas: {a[2]}")
        else:
            print("Aluno não encontrado")
    return

def guardar(turma, fnome):
    f = open(fnome, "w")
    res = ""
    for a in turma:
        nome, id, b = a
        res += str(nome) + ";" + str(id) + ";" + str(b[0]) + "::" + str (b[1]) + "::" + str(b[2]) + "|"
        res+="\n"
    f.write(res)
    f.close()

def carregar_turma(fnome):
    turma2 = []
    f = open(fnome, "r")
    linhas = f.read()
    turmat  = linhas.split("\n")

    for linha in turmat[::1]:
        t = []
        t_text=linha.split("|")

        for t_text in t_text[:-1]:
            aluno, id, b = t_text.split(";")

            b_list=[]
            for n in b.split("::"):
                b_list.append(float(n))
            a = (str(aluno), int(id), list(b_list))
            t.append(a)
        turma2.append(t)
    f.close()
    return turma2

        
    


def menu():
    print("\n========= MENU =========")
    print("1: Criar uma turma")
    print("2: Inserir um aluno na turma")
    print("3: Listar a turma")
    print("4: Consultar um aluno por ID")
    print("5: Guardar a turma em ficheiro")
    print("6: Carregar a turma dum ficheiro")
    print("0: Sair")
    print("========================")



op=-1
menu()
while op !=0:
    op=int(input("O que deseja efetuar?"))
    if  op==0:
        print("Terminou")
    elif op==1:
        turma=criar_turma()
    elif op==2:
        inserir_aluno(turma)
    elif op==3:
        listar_turma(turma)
    elif op==4:
        i = input("Qual é o id que deseja consultar?")
        consultar_id(turma,i)
    elif op==5:
        print(f"A sua turma foi guardada no ficheiro: turma")
        guardar(turma,"turma.txt")
    elif op==6:
        print(carregar_turma("turma.txt"))
    else:
        print("Tente outro número")
    

    










    
```
