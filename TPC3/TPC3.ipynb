```python
import random

def lerLista():
    num = int(input("Quantos números vai ter a sua lista?"))
    i = 0
    while i < num:
        elem = float(input("Qual vai ser o seu número?"))
        lista.append(elem)
        i = i + 1
    return lista

def criarLista():
    num = random.randint(0, 100)
    for i in range(num):
        elem = random.randint(1, 100)
        lista.append(elem)
    return lista

def soma(lista):
    soma = 0
    for elem in lista:
        soma = soma + elem
    return soma

def media (lista):
    media = soma(lista) / len(lista)
    return media

def maior(lista):
    
    maior = lista[0]
    for elem in lista[1::]:
        if maior < elem:
            maior = elem
    return maior

def menor(lista):
    menor = lista[0]
    for elem in lista[1::]:
        if menor > elem:
            menor = elem
    return menor

def decrescente(lista):
    for i in range (len(lista) - 1):
        if lista[i] < lista [i+1]:
            return print("Não")
        return print("Sim")
    
def crescente(lista):
    for i in range (len(lista) - 1):
        if lista[i] > lista[i + 1]:
            return print("Não")
        return print("Sim")
def procurarElem(lista):
    num = int(input("Insira o número que deseja procurar"))
    cond = -1
    if num in lista:
        cond = lista.index(num) + 1
    return cond

menu = -1

while menu != "0":

    menu = input(""" Qual opção deseja selecionar?  
        (1) Criar Lista 
        (2) Ler Lista  
        (3) Soma
        (4) Média
        (5) Maior
        (6) Menor
        (7) estaOrdenada por ordem crescente
        (8) estaOrdenada por ordem decrescente
        (9) Procura um elemento
        (0) Sair""")
        
    if menu == "1" :
        lista = []
        print(criarLista())

    elif menu == "2" :
        lista = []
        print(lerLista())

    elif menu == "3" :
        print(soma(lista))

    elif menu == "4":
        print(media(lista))

    elif menu == "5" : 
        print(maior(lista))

    elif menu == "6" :
        print(menor(lista))

    elif menu == "7":
        print(crescente(lista))

    elif menu == "8":
        print(decrescente(lista))

    elif menu == "9":
        print(procurarElem(lista))

print(f"A sua lista é: {lista}")




    

             



```
