```python
def listar(cinema):
    if not cinema:
        print("Não há salas no cinema.")
    else:
        print("Filmes em exibição:")
        for s in cinema:
            print(f"- {s[3]} (na {s[0]})")

def disponivel(cinema, filme, lugar):
    for s in cinema:
        if s[3] == filme:
            return lugar not in s[2]
    print("Filme não encontrado.")
    return False

def vendebilhete(cinema, filme, lugar):
    for s in cinema:
        if s[3] == filme:
            if lugar not in s[2]:
                s[2].append(lugar)
                print(f"Lugar {lugar} reservado para '{filme}' ({s[0]}).")
            else:
                print("Lugar já está ocupado.")
            return cinema
    print("Filme não encontrado.")
    return cinema

def listardisponibilidades(cinema):
    if not cinema:
        print("Não há salas registadas.")
    else:
        print("Disponibilidades:")
        for s in cinema:
            livres = s[1] - len(s[2])
            print(f"{s[0]} - '{s[3]}': {livres} lugares disponíveis")

def inserirSala(cinema, sala):
    for s in cinema:
        if s[0] == sala[0]:
            print("Essa sala já existe.")
            return cinema
    cinema.append(sala)
    print(f"Sala {sala[0]} adicionada com sucesso.")
    return cinema

def libertaLugar(cinema, filme, lugar):
    for s in cinema:
        if s[3] == filme:
            if lugar in s[2]:
                s[2].remove(lugar)
                print(f"Lugar {lugar} libertado em '{filme}'.")
            else:
                print("Esse lugar não está ocupado.")
            return cinema
    print("Filme não encontrado.")
    return cinema

def removerSala(cinema, nsala):
    for s in cinema:
        if s[0] == nsala:
            if not s[2]:
                cinema.remove(s)
                print(f"Sala {nsala} removida.")
            else:
                print("Não é possível remover — ainda há lugares ocupados.")
            return cinema
    print("Sala não encontrada.")
    return cinema

def listarSala(cinema, nsala):
    for s in cinema:
        if s[0] == nsala:
            print(f"""--- {s[0]} ---
            Filme: {s[3]}
            Lugares totais: {s[1]}
            Lugares ocupados: {s[2]}""")
            return
    print("Sala não encontrada.")

def menu():
    cinema = []

    while True:
        print(""" Gestão de Cinema 
        1. Inserir sala
        2. Listar filmes em exibição
        3. Listar disponibilidades
        4. Vender bilhete
        5. Libertar lugar
        6. Consultar sala
        7. Remover sala
        0. Sair""")

        op = input("Opção: ")

        if op == "1":
            nsala = input("Nome da sala: ")
            n_lugares = int(input("Número de lugares: "))
            filme = input("Filme em exibição: ")
            sala = [nsala, n_lugares, [], filme]
            cinema = inserirSala(cinema, sala)

        elif op == "2":
            listar(cinema)

        elif op == "3":
            listardisponibilidades(cinema)

        elif op == "4":
            filme = input("Filme: ")
            lugar = int(input("Lugar: "))
            if disponivel(cinema, filme, lugar):
                cinema = vendebilhete(cinema, filme, lugar)
            else:
                print("Lugar não disponível.")

        elif op == "5":
            filme = input("Filme: ")
            lugar = int(input("Lugar: "))
            cinema = libertaLugar(cinema, filme, lugar)

        elif op == "6":
            nsala = input("Nome da sala: ")
            listarSala(cinema, nsala)

        elif op == "7":
            nsala = input("Nome da sala: ")
            cinema = removerSala(cinema, nsala)

        elif op == "0":
            print("A encerrar sistema...")
            break

        else:
            print("Opção inválida.")

print(menu())




```
